# Let us Visualize the dataset of Iris:

#### Here we'll have to distinguish the Iris flower species from Iris-Setosa, Iris-Virginica, and Iris-Versicolor.
#### We'll do that by the help of their characteristics such as Sepal Length, Sepal Witdh, Petal Length, Petal Width.
#### We'll be using tools like pandas and numpy. To visualize the dataset we will use seaborn and matplotlib

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

iris = pd.read_csv("iris.csv")
```

```python
iris.head()
```

```python
print(iris.shape)
```

```python
print(iris.columns)
```

```python
iris["Species"].value_counts()
```

```python
iris.plot(kind = "scatter", x = "PetalLengthCm", y = 'SepalWidthCm');
plt.show()
```

```python


sns.FacetGrid(iris, hue = 'Species', size=5) \
    .map(plt.scatter, 'SepalLengthCm','SepalWidthCm') \
    .add_legend()
```

```python
plt.close()
sns.set_style("whitegrid")
sns.pairplot(iris, hue = "Species", size = 4)
plt.show()
```

```python
#this is to look at the individual feature in seaborn through Boxplot

sns.boxplot(x = "Species", y = "PetalLengthCm", data = iris)

plt.show()
```

```python
ax = sns.boxplot( data = iris, x = "Species", y = "PetalLengthCm")
ax = sns.stripplot( data = iris, x = "Species", y = "PetalLengthCm", jitter = True, edgecolor = "green" )
plt.show()
```

```python
sns.violinplot( x = "Species", y = "PetalLengthCm", data = iris, size = 5 )
plt.show()
```

```python
iris.head()
```

```python
sns.FacetGrid(iris, hue = "Species", size = 6)\
    .map(sns.kdeplot, "PetalLengthCm")\
    .add_legend()
```

```python
sns.pairplot(iris.drop("Id", axis = 1), hue = "Species" , size = 3)
plt.show()
```

```python
sns.pairplot(iris.drop("Id", axis = 1), hue = "Species", size = 3, diag_kind = "kde")
plt.show()
```

```python
iris.drop("Id", axis=1).boxplot(by= "Species", figsize=(12, 6))
```

```python
# One cool more sophisticated technique pandas has available is called Andrews Curves
# Andrews Curves involve using attributes of samples as coefficients for Fourier series
# and then plotting these

from pandas.plotting import andrews_curves
andrews_curves(iris.drop("Id", axis=1), "Species")
plt.show()
```

```python
# Another multivariate visualization technique pandas has is parallel_coordinates
# Parallel coordinates plots each feature on a separate column & then draws lines
# connecting the features for each data sample
from pandas.plotting import parallel_coordinates
parallel_coordinates(iris.drop("Id", axis=1), "Species")
```

```python
from pandas.plotting import radviz
radviz(iris.drop("Id", axis = 1), "Species")
```
